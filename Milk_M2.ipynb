## Importing Data For Data Visualization

```python
import warnings
warnings.filterwarnings('ignore')
```

### Importing packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing CSV File

```python
milk=pd.read_csv('milk.csv')
milk.head()
```

### Heat map Construction

```python
fig,ax=plt.subplots(figsize=(15,8))
sns.heatmap(milk.corr(),ax=ax,annot=True)
```

### Relation Between Odour and Turbidity

```python
p=sns.factorplot('Odour',data=milk,kind='count',hue='Turbidity')
p.set_xlabels('Odour')
p.set_ylabels('No. of Values')
p.fig.set_figheight(3)
```

### Relation Between Odour and Fat

```python
k=sns.factorplot('Odour',data=milk,kind='count',hue='Fat')
k.set_xlabels('Odour')
k.set_ylabels('No. Of Values')
k.fig.set_figheight(3)
```

### Data Classification Based on Grades

```python
p=sns.factorplot('Taste',data=milk,kind='count',hue='Fat')
p.set_xlabels('Taste')
p.set_ylabels('No. Of Values')
p.fig.set_figheight(3)
```

```python
milk['Grade'].hist(figsize=(10,4),color='green')
plt.title('Grades of milk')
plt.xlabel('Grades')
plt.ylabel('No. of data')
plt.show()
```

### Milk Colour Distribution

```python
milk["Colour"].hist(figsize=(5,4), color = 'blue')
plt.title("Colour distribution")
plt.xlabel('Colour')
plt.ylabel("Total Number")
plt.show()
```

```python
def Contract(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(6,6), autopct='%3f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 10)

Contract(milk, 'Grade')
```
